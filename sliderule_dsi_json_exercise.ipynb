## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# List major theme names

- Extract 'mjtheme-name' column that lists theme codes and names for each project (noting that most projects have more than one theme)
- Use json_normalize to convert the nested JSON information to an expanded data frame

```python
x = wb_df_1['mjtheme_namecode']
xdf = json_normalize(wb_df_1['mjtheme_namecode'].iloc[0])
xdf = xdf.drop(xdf.index[:])

for row in x:   
    a = json_normalize(row)
    xdf = xdf.append(a)
    row = next(x.iteritems())
    
xdf.head(20)
```

Fill missing values

```python
xdf1 = xdf.reset_index()
xdf1.head(10)
```

```python
combinations = xdf1.groupby(['code', 'name']).count()
combinations = combinations.reset_index()
combinations
```

Construct a dictionary of code-name pairs to ensure missing names are filled correctly

```python
namescodes = {1: 'Economic management', 2: 'Public sector governance', 3: 'Rule of law', 4: 'Financial and private sector development',
             5: 'Trade and integration', 6: 'Social protection and risk management', 7: 'Social dev/gender/inclusion',
             8: 'Human development', 9: 'Urban development', 10: 'Rural development', 11: 'Environment and natural resources management'}
```

```python
namescodes
```

```python
codes = np.asarray(xdf1.code)
nameslist=[]
for c in codes:
    intcode = int(c)
    name = namescodes.get(intcode)
    nameslist.append(name)
```

Add new column with all names filled from dictionary and drop column with missing names

```python
xdf1['filled_name'] = nameslist
xdf1 = xdf1.drop('name', axis=1)
xdf1.head(10)
```
